```python
import cv2
import time
import numpy as np
from ultralytics import YOLO
from norfair import Detection, Tracker

# Load YOLO model
model = YOLO("/content/drive/MyDrive/niel_deep/runs/detect/train/weights/best.pt")

# Open video
cap = cv2.VideoCapture("/content/drive/MyDrive/niel_deep/videotest1.mp4")

# Ambil resolusi asli dari video
# Karena akan di-rotate 90 derajat, tukar width dan height
original_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
original_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

frame_width = original_height
frame_height = original_width
fps_video = int(cap.get(cv2.CAP_PROP_FPS))

# Hitung ukuran baru (90% dengan menjaga rasio)
scale_percent = 150
output_width = int(frame_width * scale_percent / 100)
output_height = int(frame_height * scale_percent / 100)

# Output setup dengan ukuran baru
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('output_video_1.mp4', fourcc, fps_video, (output_width, output_height))

# Tracker
tracker = Tracker(distance_function="euclidean", distance_threshold=30)

# Hitung FPS & objek
counter, fps, elapsed = 0, 0, 0
start_time = time.perf_counter()
frames = []
class_names = ['Minuman Botol', 'Minuman Gelas', 'Minuman Kaleng', 'Minuman Kotak']
class_object_counts = {name: set() for name in class_names}

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # ROTASI FRAME JIKA VIDEO TERLIHAT TERBALIK 90 DERAJAT KE KIRI
    frame = cv2.rotate(frame, cv2.ROTATE_90_CLOCKWISE)

    # Konversi warna
    og_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    frame = og_frame.copy()

    # Predict
    results = model(frame, device=0, classes=[0, 1, 2, 3], conf=0.2)

    detections = []
    object_infos = []

    for result in results:
        boxes = result.boxes
        if not boxes:
            continue

        cls = boxes.cls.tolist()
        xyxy = boxes.xyxy.cpu().numpy()
        conf = boxes.conf.cpu().numpy()

        for box, class_id, confidence in zip(xyxy, cls, conf):
            x1, y1, x2, y2 = box
            cx = (x1 + x2) / 2
            cy = (y1 + y2) / 2
            detections.append(Detection(points=np.array([cx, cy]), scores=np.array([confidence])))
            object_infos.append((int(class_id), (x1, y1, x2, y2)))

    tracked_objects = tracker.update(detections=detections)

    for det, obj in zip(tracked_objects, object_infos):
        class_id, (x1, y1, x2, y2) = obj
        track_id = det.id
        class_name = class_names[class_id]

        # Tambahkan track ID ke kelas yang sesuai
        class_object_counts[class_name].add(track_id)

        # Visualisasi
        color_id = track_id % 3
        color = [(0, 0, 255), (255, 0, 0), (0, 255, 0)][color_id]
        cv2.rectangle(og_frame, (int(x1), int(y1)), (int(x2), int(y2)), color, 2)
        cv2.putText(og_frame, f"{class_name}-{track_id}", (int(x1)+10, int(y1)-5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 0), 1)

    # Hitung FPS
    current_time = time.perf_counter()
    elapsed = current_time - start_time
    counter += 1
    if elapsed > 1:
        fps = counter / elapsed
        counter = 0
        start_time = current_time

    # Tampilkan count tiap kelas
    y_text = 30
    for class_name, track_ids in class_object_counts.items():
        count = len(track_ids)
        cv2.putText(og_frame, f"{class_name}: {count}", (10, y_text),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
        y_text += 25

    # Resize dengan aspect ratio tetap
    resized_frame = cv2.resize(og_frame, (output_width, output_height), interpolation=cv2.INTER_AREA)
    out.write(cv2.cvtColor(resized_frame, cv2.COLOR_RGB2BGR))
    frames.append(resized_frame)

cap.release()
out.release()
```
